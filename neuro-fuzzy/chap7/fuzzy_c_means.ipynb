```python
from fuzzycmeans.fuzzy_clustering import FCM
from fuzzycmeans.visualization import draw_model_2d
import logging
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

## データを読み込む

```python
dataset = pd.read_csv("AirlinesCluster.csv")
dataset = dataset[["Balance", "BonusMiles"]][:500]
```

データの標準化

```python
dataset_standardized = pd.DataFrame(preprocessing.scale(dataset))
```

## modelを作る

```python
fcm = FCM(n_clusters=5)
fcm.set_logger(tostdout=False)
```

学習する

```python
fcm.fit(dataset_standardized)
```

予測する

```python
predicted_membership = fcm.predict(np.array(dataset_standardized))
```

結果を見てみる

```python
draw_model_2d(fcm, data=np.array(dataset_standardized), membership=predicted_membership)
```

```python
import skfuzzy as sk
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_qual = np.arange(0, 11, 1)

qual_lo = sk.trimf(x_qual, [0, 0, 5])
qual_md = sk.trimf(x_qual, [0, 5, 10])

tri_pro = sk.cartprod(qual_lo, qual_md)
```

```python
plt.subplot(211)
plt.plot(x_qual, qual_lo)
plt.plot(x_qual, qual_md)

plt.subplot(212)
for i in range(len(tri_pro)): 
    plt.plot(x_qual, tri_pro[i]) 
```

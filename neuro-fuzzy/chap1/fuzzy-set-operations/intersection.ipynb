```python
import skfuzzy as sk
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_qual = np.arange(0, 11, 1)

qual_lo = sk.trimf(x_qual, [0, 0, 5])
qual_hi = sk.trimf(x_qual, [0, 5, 10])

tri_and = sk.fuzzy_and(x_qual, qual_lo, x_qual, qual_hi)
```

```python
plt.plot(x_qual, qual_lo)
plt.plot(x_qual, qual_hi)
plt.plot(tri_and[0], tri_and[1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 12)
```

## Sigmoid Activation Function
$Sigmoid(x) = \frac{1}{1 + e^{-x}}$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
sigmoid_output = sigmoid(x)
```

```python
plt.figure()
plt.plot(x, sigmoid_output)
```

## Tanh Actibation Function
$tanh(x) = \frac{2}{1 + e^{-2x}}$

```python
def tanh(x):
    return np.tanh(x)
```

```python
tanh_output = tanh(x)
```

```python
plt.figure()
plt.plot(x, tanh_output)
```

## Softmax Activation Function
$softmax(x) = \frac{e^{i}}{\sum e^{i}}$

```python
def softmax(x):
    exps = np.exp(x)
    return exps / np.sum(exps)
```

```python
softmax_output = softmax(x)
```

```python
plt.figure()
plt.plot(x, softmax_output)
```

## ReLU Activation Function
$ReLU(x) = max(0, x)$

```python
def relu(x):
    return x*(x > 0)
```

```python
relu_output = relu(x)
```

```python
plt.figure()
plt.plot(x, relu_output)
```

## Leaky ReLU Activation Function
$LeakyReLU(x) = \{x, if x > 0\}$

$\{\alpha x, otherwise\}$

```python
def leaky_relu(x, func, *args):
    return np.where(x <= 0, func(x, alpha), x)
```

```python
func = lambda x, alpha: alpha * (np.exp(x) - 1)
alpha = 0.5
leaky_relu_output = leaky_relu(x, func, alpha)
```

```python
plt.figure()
plt.plot(x, leaky_relu_output)
```

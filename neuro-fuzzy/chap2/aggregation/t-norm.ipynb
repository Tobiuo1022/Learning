```python
import numpy as np
import skfuzzy as sk
import matplotlib.pyplot as plt
%matplotlib inline
```

## 関数を定義する

```python
def t_norm(mfx, mfy):
    tnorm = np.fmin(mfx, mfy)
    return tnorm
```

```python
x = np.arange(0, 80, 0.1)

full_speed = sk.sigmf(x, 80, 2)
slow = sk.sigmf(x, 30, 2)

tnorm = t_norm(full_speed, slow)
```

```python
plt.plot(x, tnorm)
```

## 3次元でplotしてみる

```python
X, Y = np.meshgrid(full_speed, slow)
tnorm3D = t_norm(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.set_xlabel("full speed")
ax.set_ylabel("slow")
ax.view_init(azim=-120)
ax.plot_surface(X, Y, tnorm3D)
cs = ax.contourf(X,Y,tnorm3D,100, cmap="RdBu")
plt.colorbar(cs)
```
